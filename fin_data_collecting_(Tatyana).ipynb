```python
def check_company(fin):
  year_2017 = 0
  year_2018 = 0
  year_2019 = 0
  for report in fin['data']:
    if (report['startDate'] >= '2017-01-01' and report['startDate'] <= '2017-12-31') or (report['endDate'] >= '2017-01-01' and report['endDate'] <= '2017-12-31'):
      year_2017 = 1
    if (report['startDate'] >= '2018-01-01' and report['startDate'] <= '2018-12-31') or (report['endDate'] >= '2018-01-01' and report['endDate'] <= '2018-12-31'):
      year_2018 = 1
    if (report['startDate'] >= '2019-01-01' and report['startDate'] <= '2019-12-31') or (report['endDate'] >= '2019-01-01' and report['endDate'] <= '2019-12-31'):
      year_2019 = 1
  if year_2017 == 1 and year_2018 == 1 and year_2019 == 1:
    return True
  return False
```

### Financial reports' parsing

```python
def get_report(fin):
  financial_statements_dataset = pd.DataFrame()
  financial_statements_dataset.at[0, 'Ticker'] = fin['data'][0]['symbol']
  for report in fin['data']:
      if (report['year'] == 2017):
        for variable  in report['report']['bs']:
          if variable['concept'] == 'us-gaap_Assets':
            financial_statements_dataset['Total assets 2017'] = variable['value']
          if variable['concept'] == 'us-gaap_StockholdersEquity':
            financial_statements_dataset['Total shareholders\' equity 2017'] = variable['value']
        for variable  in report['report']['ic']:
          if variable['concept'] == 'us-gaap_NetIncomeLoss':
            financial_statements_dataset['Net income 2017'] = variable['value']
          if variable['concept'] == 'us-gaap_OperatingIncomeLoss':
            financial_statements_dataset['Operating income 2017'] = variable['value']

      if (report['year'] == 2018):
        for variable  in report['report']['bs']:
          if variable['concept'] == 'us-gaap_Assets':
            financial_statements_dataset['Total assets 2018'] = variable['value']
          if variable['concept'] == 'us-gaap_StockholdersEquity':
            financial_statements_dataset['Total shareholders\' equity 2018'] = variable['value']
        for variable  in report['report']['ic']:
          if variable['concept'] == 'us-gaap_NetIncomeLoss':
            financial_statements_dataset['Net income 2018'] = variable['value']
          if variable['concept'] == 'us-gaap_OperatingIncomeLoss':
            financial_statements_dataset['Operating income 2018'] = variable['value']

      if (report['year'] == 2019):
        for variable  in report['report']['bs']:
          if variable['concept'] == 'us-gaap_Assets':
            financial_statements_dataset['Total assets 2019'] = variable['value']
          if variable['concept'] == 'us-gaap_StockholdersEquity':
            financial_statements_dataset['Total shareholders\' equity 2019'] = variable['value']
        for variable  in report['report']['ic']:
          if variable['concept'] == 'us-gaap_NetIncomeLoss':
            financial_statements_dataset['Net income 2019'] = variable['value']
          if variable['concept'] == 'us-gaap_OperatingIncomeLoss':
            financial_statements_dataset['Operating income 2019'] = variable['value']

  return financial_statements_dataset
```
